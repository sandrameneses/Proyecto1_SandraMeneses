1. Extraccion archivo user_reviews

```python
#importo librerias

! pip list
import json
import pandas as pd
import ast
import gzip

# renombro el archivo comprimido
archivo_reviews = "/user/Documents/SANDRA 2023/HENRY/PROYECTO 1 RESPALDO/PI MLOps - STEAM/user_reviews.json.gz"

# creo lista vacia
lista = []

# se abre el archivo en modo lectura, se itera y evalua linea por linea y se agrega a la lista
with gzip.open(archivo_reviews, "rt", encoding= "utf8") as archivo:
        for line in archivo.readlines():
            lista.append(ast.literal_eval(line))

```

```python
# creo a partir de la lista Dataframe y reviso su forma

df = pd.DataFrame(lista)
df
```

```python
#reviso los tipos de datos en cada columna del dataframe

df.dtypes
```

```python
# creo un nuevo dataframe con las columnas del archivo

Revision_usuarios = df [["user_id", "user_url", "reviews"]]
```

```python
# convierto el archivo en csv y lo guardo

Revision_usuarios.to_csv("/user/Documents/SANDRA 2023/HENRY/PROYECTO 1/DATA EXTRACCION/revision_usuarios.csv", index = False)
```

2. Extraccion archivo steam_games
